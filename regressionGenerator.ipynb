Linear Regression: Generates a Dataframe with a regression bias

```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

import random
import math
```

```python
def regressionGenerator(y_var, x_var, positive = True, n_samples=100, dummy=None):
    data = {}
    cont = np.random.randint(20, 66, n_samples)
    
    if positive:
        slope = 3
    else:
        slope = -3
    
    y_line = cont*slope

    if positive:
        int = 10
    else:
        int = np.abs(np.min(y_line)) + 10

    depen = int +  y_line + np.random.normal(0, 20, n_samples)
    data[y_var]= depen
    data[x_var] = cont
    if dummy:
        data[dummy] = np.random.randint(0, 2, n_samples)

    return pd.DataFrame(data)
```

```python
y_v = "wage"
cont_v = "age"
dummy_v = "gender"
df = regressionGenerator(y_v, cont_v, positive = False, dummy = dummy_v)
# plt.scatter(df[cont_v], df[y_v], c=df[dummy_v]) # plotting single point
df
```

Clustering: Generates a Dataframe with a clustering bias

```python
# Returns dataset for clustering
def getClustering(numClasses = 3, rangeClasses = 1000, rangeInstances = 150, numInstancesPerClass = 50):
    datapointsX = []
    datapointsY = []

    classCentersPresentation = [(600,250), (200,650), (800,800)]

    for i in range(numClasses):
        classCenterX = random.randint(0,rangeClasses)
        classCenterY = random.randint(0,rangeClasses)

        # FOR PRESENTATION
        classCenterX = classCentersPresentation[i][0]
        classCenterY = classCentersPresentation[i][1]

        for _ in range(random.randint(numInstancesPerClass*2//3,numInstancesPerClass)):
            R = rangeInstances
            theta = random.uniform(0,2*math.pi)
            r = random.uniform(0,R)
            datapointsX.append(classCenterX + math.sqrt(r*R)*math.cos(theta))
            datapointsY.append(classCenterY + math.sqrt(r*R)*math.sin(theta))

    dict = {'x':datapointsX, 'y':datapointsY}
    df = pd.DataFrame(dict)

    return df
```

```python
df = getClustering()
plt.scatter(df['x'], df['y'])
```
