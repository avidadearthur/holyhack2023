```python
import math
import random
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Returns dataset for clustering
def getClustering():
    numClasses = 3
    rangeClasses = 1000
    rangeInstances = 150
    numInstancesPerClass = 50

    datapointsX = []
    datapointsY = []

    classCentersPresentation = [(600,250), (200,650), (800,800)]

    for i in range(numClasses):
        classCenterX = random.randint(0,rangeClasses)
        classCenterY = random.randint(0,rangeClasses)

        # FOR PRESENTATION
        classCenterX = classCentersPresentation[i][0]
        classCenterY = classCentersPresentation[i][1]

        for _ in range(random.randint(numInstancesPerClass*2//3,numInstancesPerClass)):
            R = rangeInstances
            theta = random.uniform(0,2*math.pi)
            r = random.uniform(0,R)
            datapointsX.append(classCenterX + math.sqrt(r*R)*math.cos(theta))
            datapointsY.append(classCenterY + math.sqrt(r*R)*math.sin(theta))

    dict = {'x':datapointsX, 'y':datapointsY}
    df = pd.DataFrame(dict)

    return df

df = getClustering()

df   
```

```python
plt.scatter(df['x'], df['y'])
```

```python
def lsvmClassification():
    X = []
    Y = []
    numPoints = random.randint(250,400)
    rang = 100

    # f = 2.3x + 23
    factor = 0.65
    add = 15
    separation = 15

    for i in range(numPoints):
        x = random.uniform(0,rang)
        y = x*factor+add
        while y < x*factor+add+separation and y > x*factor+add-separation:
            y = random.uniform(0,rang)
        
        X.append(x)
        Y.append(y)

    dict = {'x':X, 'y':Y}
    df = pd.DataFrame(dict)

    return(df)
       

df = lsvmClassification()
df
```

```python
plt.scatter(df['x'], df['y'])
```

```python
def getLinearRegression():
    n_samples = 100
    positive = True
    y_var = "y"
    x_var = "x"
    dummy = "gender"

    data = {}
    cont = np.random.randint(20, 66, int(n_samples))


    if positive:
        slope = 3
    else:
        slope = -3

    y_line = cont*slope

    if positive:
        inter = 10
    else:
        inter = np.abs(np.min(y_line)) + 10

    depen = inter +  y_line + np.random.normal(0, 20, n_samples)
    data[y_var]= depen
    data[x_var] = cont
    if dummy:
        data[dummy] = np.random.randint(0, 2, n_samples)
    df = pd.DataFrame(data)
    return df

df = getLinearRegression()
df
```
